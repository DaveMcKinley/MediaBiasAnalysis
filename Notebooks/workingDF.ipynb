```python
df3 = pd.read_csv("data/articles3.csv")
df3.head()
```

```python
df3.shape
```

```python
df3["publication"].value_counts()
```

```python
df3 = df3.drop(drop, axis = 1)
df3
```

```python
vox = pd.read_pickle("data/vox.pkl")
vox.head()
```

```python
drop = [0, 1, 2, 3, 4, 5, 6, 7, 9, 10]
vox= vox.drop(drop, axis = 1)
vox.head()
```

```python
vox = vox[[11, 8]]
```

```python
vox = vox.rename(columns = {11:"publication", 8: "content"})
vox
```

```python
cnn = df1[df1["publication"] == "CNN"]
cnn
```

```python
left = pd.concat([cnn, vox])
left = left.rename(columns={"publication": "Publication", "content": "Content"})
left
```

```python
left.info()
```

```python
left["Bias"] = "Left"
left.reset_index(drop = True)
```

```python
hill = pd.read_pickle("data/hill.pkl")
hill.head()
```

```python
drop = [0, 1, 2, 3, 4, 5, 6, 7, 9, 10, "index"]
hill = hill.drop(drop, axis = 1)
hill
```

```python
hill = hill[[11, 8]]
```

```python
hill = hill.rename(columns = {11:"publication", 8: "content"})
hill
```

```python
hill.info()
```

```python
reuters =  df3[df3["publication"] == "Reuters"]
reuters
```

```python
center = pd.concat([reuters, hill])
center = center.rename(columns={"publication": "Publication", "content": "Content"})
center
```

```python
center["Bias"] = "Center"
center.reset_index(drop = True)
```

```python
breitbart = df1[df1["publication"] == "Breitbart"]
breitbart
```

```python
review = df2[df2["publication"] == "National Review"]
review
```

```python
right = pd.concat([breitbart, review])
right = right.rename(columns={"publication": "Publication", "content": "Content"})
right
```

```python
right["Bias"] = "Right"
right.reset_index(drop = True)
```

```python
df = pd.concat([left, center, right]).reset_index(drop = True)
df
```
